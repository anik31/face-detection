```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = plt.imread('./pic.jpg')
```

```
plt.imshow(img)
```

```
detector = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```
faces = detector.detectMultiScale(img)
```

```
faces
```

```
faces.shape
```

```
x,y,w,h=faces[0]
```

```
img1=cv2.rectangle(img, (x,y),(x+w,y+h),[255,255,255],2)
```

```
plt.imshow(img1)
```

```
for face in faces:
  x,y,w,h=face
  img1=cv2.rectangle(img, (x,y),(x+w,y+h),[255,255,255],2)
```

```
plt.imshow(img1)
```
